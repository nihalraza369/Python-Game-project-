```python
import time
import random

print("‚è±Ô∏è Welcome to the Reaction Time Tester Game!")
print("Get ready... When you see 'Now!', press Enter as fast as you can.")
print("Type 'exit' anytime to quit.")
print("-" * 50)

while True:
    ready = input("\nType 'go' to start or 'exit' to quit: ").lower()

    if ready == "exit":
        print("üëã Exiting game. Stay sharp!")
        break
    elif ready != "go":
        print("‚ö†Ô∏è Type 'go' to begin.")
        continue

    wait_time = random.randint(2, 5)
    print("‚è≥ Wait for it...")
    time.sleep(wait_time)

    print("üëâ Now!")
    start_time = time.time()
    input()  # Wait for user to press Enter
    end_time = time.time()

    reaction_time = round(end_time - start_time, 3)
    print(f"‚ö° Your reaction time: {reaction_time} seconds")
```
